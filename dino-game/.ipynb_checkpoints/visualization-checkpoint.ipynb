```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel('two_hidden_layers_sigmoid.xlsx')
```

```python
x = data['Generation']
y = data['Average']
```

```python
plt.plot(x, y)

plt.title("Trend Line")
plt.xlabel("Generation")
plt.ylabel("Average Score")

plt.show()

z = data['Best Score']
plt.plot(x, z)
plt.title("Trend Line")
plt.xlabel("Generation")
plt.ylabel("Best Score")

plt.show()
```
